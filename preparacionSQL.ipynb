```python
# SQLite3

import sqlite3
import pandas as pd

# Crear conexión a base de datos SQLite en memoria
conn = sqlite3.connect(':memory:')
```

```python
# Get a cursor object from the connection
cursor = conn.cursor() # Get a new cursor associated with this connection.

# Crear tablas y poblar datos
# Check if table exists before creating
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='empleados'")
table_exists = cursor.fetchone()

if table_exists:
    # Drop the table if it exists
    cursor.execute("DROP TABLE empleados")
    conn.commit()
    print("Table 'empleados' dropped.")  # Print a message for clarity

cursor.execute('''
CREATE TABLE empleados (
    id INTEGER PRIMARY KEY,
    nombre TEXT,
    departamento TEXT,
    salario REAL,
    fecha_ingreso TEXT
);
''')

empleados_data = [
    (1, 'Ana Gómez', 'Ventas', 3000.0, '2020-05-15'),
    (2, 'Luis Rojas', 'Marketing', 3500.0, '2019-03-10'),
    (3, 'Sofía Pérez', 'Ventas', 2800.0, '2021-07-01'),
    (4, 'Carlos Díaz', 'IT', 4000.0, '2018-11-21'),
    (5, 'Lucía Fernández', 'IT', 4200.0, '2017-06-30'),
]

cursor.executemany("INSERT INTO empleados VALUES (?, ?, ?, ?, ?);", empleados_data)
conn.commit()
```

```python
# Seleccionar todos los empleados
query = "SELECT * FROM empleados;"
df = pd.read_sql_query(query, conn)
print("Todos los empleados:")
display(df)

# Seleccionar nombre y salario
query = "SELECT nombre, salario FROM empleados;"
df = pd.read_sql_query(query, conn)
display(df)

# Filtrar por departamento
query = "SELECT * FROM empleados WHERE departamento = 'IT';"
df = pd.read_sql_query(query, conn)
display(df)
```

```python
# Ordenar por salario descendente
query = "SELECT nombre, salario FROM empleados ORDER BY salario DESC;"
df = pd.read_sql_query(query, conn)
display(df)

# Agrupar por departamento y calcular promedio
query = """
SELECT departamento, AVG(salario) as salario_promedio
FROM empleados
GROUP BY departamento;
"""
df = pd.read_sql_query(query, conn)
display(df)
```

```python
# Subconsulta: empleados con salario mayor al promedio
query = """
SELECT nombre, salario FROM empleados
WHERE salario > (SELECT AVG(salario) FROM empleados);
"""
df = pd.read_sql_query(query, conn)
display(df)
```
